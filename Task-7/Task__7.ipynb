# Problem 1: Smart Light Controller

```python
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"
    def turn_on(self):
        self.status = "ON"
    def turn_off(self):
        self.status = "OFF"

    def display_status(self):
        print(f"{self.name} is {self.status}")

light_name =input("Light Name: ")
action =input("Action (ON/OFF): ").upper()
light =SmartLight(light_name)
if action =="ON":
    light.turn_on()
elif action =="OFF":
    light.turn_off()
else:
    print("Invalid action")
light.display_status()
```

# Problem 2: Employee ID Card System

```python
class Employee:
    def __init__(self, name, emp_id, department):
        self.name =name
        self.emp_id =emp_id
        self.department =department

    def display_id_card(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")

name =input("Employee Name: ")
emp_id =input("Employee ID: ")
department =input("Department: ")
emp =Employee(name,emp_id,department)
emp.display_id_card()
```

# Problem 3: Mobile Contact Record

```python
class Contact:
    def __init__(self, name, phone):
        self.name =name
        self.phone =phone
    def display_contact(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")

name = input("Contact Name: ")
phone = input("Phone Number: ")
contact = Contact(name, phone)
contact.display_contact()
```

# Problem 4: Product Price Tag Generator

```python
class Product:
    def __init__(self, name, price):
        self.name =name
        self.price =price
    def display_price_tag(self):
        print(f"Product: {self.name}")
        print(f"Price: ₹{self.price}")

name = input("Product Name: ")
price = input("Price: ")
product = Product(name, price)
product.display_price_tag()
```

# Problem 5: Movie Rating Display System

```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating
    def display_movie(self):
        print(f"Movie: {self.name}")
        print(f"Rating: {self.rating} / 5")

name = input("Movie Name: ")
rating = input("Rating: ")
movie = Movie(name, rating)
movie.display_movie()
```

# Problem 6: Delivery Address Manager

```python
class Delivery:
    def __init__(self, customer_name, address):
        self.customer_name =customer_name
        self.address =address
    def display_delivery(self):
        print("Delivery Details")
        print(f"Customer: {self.customer_name}")
        print(f"Address: {self.address}")

customer_name =input("Customer Name: ")
address =input("Address: ")
delivery = Delivery(customer_name, address)
delivery.display_delivery()
```
