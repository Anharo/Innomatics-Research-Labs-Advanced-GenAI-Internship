# Problem Statement 1: Smart Parking Lot Management System

```python
def parking_system(capacity,logs):
  parked=0
  peak =0
  for log in logs:
    if log=="IN":
      parked+=1
    elif log=="OUT":
      if parked>0:
        parked-=1
    if parked>peak:
      peak=parked

  print("Currently Parked Vehicles: ",parked)
  print("Peak Parking Usage: ",peak)

  if parked>=capacity:
    print("Parking Status: ALERT! Exceeds Capacity")
  else:
    print("Parking Status: Available")

capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
parking_system(capacity,vehicle_logs)
```

 # Problem Statement 2: Online Food Delivery Time Estimator

```python
def estimator(distance,traffic,weather):
  estimate_time = distance*2
  if traffic =="Low":
    estimate_time +=5
  elif traffic =="Medium":
    estimate_time +=10
  elif traffic =="High":
    estimate_time +=20

  if weather =="Cloudy":
    estimate_time +=5
  elif weather =="Rainy":
    estimate_time +=15
  print("Estimated Delivery Time:", estimate_time, "minutes")

distance=8
traffic="High"
weather="Rainy"
estimator(distance, traffic, weather)
```

 # Problem Statement 3: Movie Theatre Seat Occupancy Analyzer

```python
def occupancy_analyze(total,booked):
  total_booked = len(booked)
  occupancy = (total_booked/total)*100

  if occupancy ==100:
    status= "HouseFull"
  elif occupancy>=75:
    status ="Almost Full"
  else:
    status ="Seats Available"

  if occupancy>=90:
    suggestion= "Suggestion: Open additional shows"
  else:
    suggestion= "Suggestion: No additional shows needed"

  print("Occupancy:", int(occupancy), "%")
  print("Show Status:", status)
  print(suggestion)

total_seats =200
booked_seats =[1]*150
occupancy_analyze(total_seats, booked_seats)
```

# Problem Statement 4: Cloud Server Load Classification System

```python
def server_load(readings):
  total=0
  for reading in readings:
    total+=reading
  average=total/len(readings)
  if average<50:
    status="Normal"
  elif average>=50 and average<=80:
    status="Warning"
  else:
    status="Critical"
  print("Average CPU Load:", int(average),"%")
  print("Server Status:", status)

cpu_readings = [45, 60, 70, 85, 90]
server_load(cpu_readings)
```

# Problem Statement 5:Smart Classroom Resource Usage Monitor


```python
def resource_usage(usage):
  overuse=[]
  for resoruce ,time in usage.items():
    if time>8:
      overuse.append(resoruce)
  if overuse:
    print("Overused Resource: ",",".join(overuse))
    print("Energy Alert: Yes")
  else:
    print("No Overused Resources")
    print("Energy Alert: No")

usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}
resource_usage(usage)
```

# Problem Statement 6: Online Event Registration Capacity Controller


```python
def registrations(capacity,regis):
  confirmed=0
  waitlist= 0
  for i in range(regis):
    if confirmed<capacity:
      confirmed+=1
    else:
      waitlist+=1
  if confirmed>=capacity:
    status= "Closed"
  else:
    status= "Open"

  print("Confirmed Registrations: ",confirmed)
  print("Waitlist Registrations: ",waitlist)
  print("Registration Status: ",status)

capacity =100
regis=105
registrations(capacity, regis)
```
