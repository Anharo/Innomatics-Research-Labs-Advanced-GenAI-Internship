# 1. User Login Check

```python
# Declaring Default username and password
username = "admin"
password = "1234"
# User Input for username and password
user_input = input("Enter username:")
password_input = input("Enter password:")
# If condition to check if username and password matches
if user_input == username and password_input == password:
    print("Login Successful")
else:
  print("Invalid Credentials")
```

# 2. Pass / Fail Analyzer

```python
# Defining a list of marks
marks = [45, 78, 90, 33, 60]
pass_count= 0
fail_count= 0
# Checking if marks are greater than 50 one at a time
for i in marks:
    if i>=50:
        pass_count+= 1
    else:
        fail_count+= 1
# Clearly printing Total Students,Passed,Failed.
print("Total Students:", len(marks))
print("Total Pass Students:", pass_count)
print("Total Fail Students:", fail_count)
```

# 3. Simple Data Cleaner

```python
names = [" Alice ", "bob", " CHARLIE "]
# Using list Comprehension to strip the whitespaces from front and end and .lower() to decapitalize the words.
cleaned_names = [name.strip().lower() for name in names]
print("Cleaned Names:", cleaned_names)
```

# 4. Message Length Analyzer

```python
messages = ["Hi", "Welcome to the platform", "OK"]
# Checking each word one by one then printing all Message Length Ether short or long clearly
for message in messages:
    length = len(message)
    print(f"Message: {message}")
    print(f"Length: {length}")
    if length > 10:
        print("Status: Long message")
    else:
        print("Status: Short message")
    print("-" * 30) #Used -- to seperate Different message and there details
```

# 5. Error Message Detector

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count = 0 # Defining a counter
for log in logs:
    if log == "ERROR": #Checks if the log is Error or not
        error_count += 1 #Increment the counter

print("Total ERROR entries:", error_count)
```
