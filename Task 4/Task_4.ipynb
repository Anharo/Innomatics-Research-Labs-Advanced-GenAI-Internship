# Problem 1: Mobile Recharge Validation System

```python
def valid_recharge(amount):
  valid_plans=[199,299,399,599]
  # Checking if the value is in valid plans or not
  if amount<50:
      print("Recharge amount must be at least ₹50.")
      return False
  if amount not in valid_plans:
      print("Invalid plan selected.")
      print("Valid plans are:", valid_plans)
      return False

  return True
# Loop until the input  is correct
while(True):
      try:
        amount=int(input("Enter recharge amount: ₹"))
        if valid_recharge(amount):
            print("Recharge successful!")
            break
        else:
            print("Please try again.\n")
      except ValueError:
        print("Please enter a valid number.\n")
```

# Problem 2: Inventory Reorder Alert System


```python
products = {
    "Mobile": 12,
    "Laptop": 25,
    "Headphones": 8,
    "Keyboard": 18,
    "Mouse": 14
}

def check_stock(products):
    for product,quantity in products.items():
        if quantity<15:
            print(f"{product}:Reorder Alert (Stock = {quantity})")
        else:
            print(f"{product}:Stock OK (Stock = {quantity})")
check_stock(products)
```

# Problem 3: Student Result Processing System

```python
student_marks = [60,45,70,55,40]
def check_result(marks):
    total=0
    # Calculate total marks
    for mark in marks:
        total+=mark
    # Finding Average
    average=total/len(marks)
    print("Average marks:", average)
    # Checking if marks is greater than average or not
    if average>=50:
        return "Pass"
    else:
        return "Fail"

result=check_result(student_marks)
print("Result:", result)
```

# Problem 4: Cab Fare Estimator with Retry Option

```python
def calculate_fare(distance, peak_hour):
    base_fare=50
    per_km_rate=12
    # calculate normal fare
    fare=base_fare+(distance*per_km_rate)
    # add peak hour extra
    if peak_hour.lower()=="yes":
        fare=fare+(fare*0.25)

    return fare
# retry system using while loop
while True:
    try:
        distance=float(input("Enter distance in km: "))
        if distance<=0:
            print("Distance must be greater than 0.\n")
            continue
        peak_hour=input("Is it peak hour? (yes/no): ")
        fare=calculate_fare(distance,peak_hour)
        print(f"Total Fare: ₹{fare:.2f}")
        break
    except ValueError:
        print("Please enter valid input.\n")
```

# Problem 5: Employee Attendance Eligibility Checker

```python
attendance_list = ["P","A","P","P","A","P","P","P"]
def check_attendance(attendance):
    present_count=0
    # loop to count present days
    for day in attendance:
        if day=="P":
            present_count+=1

    # calculate percentage
    percentage=(present_count/len(attendance))*100
    print(f"Attendance Percentage: {percentage:.2f}%")
    if percentage>=75:
        return "Eligible"
    else:
        return "Not Eligible"
# Calling the function and saving the result in result variable
result = check_attendance(attendance_list)
print("Status:", result)
```

# Problem 6: Password Strength Checker

```python
def check_password_strength(password):
    digit =False
    special =False
    special_chars ="@#$"
    # check minimum length of the password
    if len(password)<8:
        return "Weak Password (Minimum length is 8)"

    # loop over string to check strength
    for char in password:
        if char.isdigit():
            digit=True
        if char in special_chars:
            special=True

    # check conditions
    if digit and special:
        return "Strong Password"
    else:
        return "Weak Password (Must contain digit and special character)"
password=input("Enter password: ")
result=check_password_strength(password)
print(result)
```
