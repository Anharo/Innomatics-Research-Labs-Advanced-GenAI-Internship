# Problem 1: Unique Words in a Sentence

```python
# Take input from user
s=input("Enter sentence: ")
# Split sentence into words
words=s.split()
# Convert list to set to remove duplicates
unique=set(words)
print("Unique words count:", len(unique))
print("Unique words:", unique)
```

# Problem 2: Highest Salary from Employee Data


```python
# Dictionary of employees and salaries
emp= {"Ravi": 75000,"Anita": 68000,"Kiran": 72000}
# Finding employee with highest salary
high_emp = max(emp, key=emp.get)
print("Highest Salary:", high_emp, "-", emp[high_emp])
```

# Problem 3: Find Maximum and Minimum Values

```python
# Given list
numbers = [45, 22, 89, 10, 66]

# Initialize max and min with first element
max_num = numbers[0]
min_num = numbers[0]

# Comparing each element using loop
for i in numbers:
    if i>max_num:
        max_num= i
    if i< min_num:
        min_num= i

print("List:", numbers)
print("Max:", max_num)
print("Min:", min_num)
```

# Problem 4: Count Products Above a Price Threshold

```python
# List of product prices
prices = [450, 1200, 899, 1500, 300]
# Initialize counter
count = 0
# Loop through prices with check condition for prices above 1000
for price in prices:
    if price > 1000:
        count += 1
print("Products above 1000:", count)
```

# Problem 5: Calculate Attendance Percentage

```python
# Attendance list
attendance = ["P", "P", "A", "P", "P"]
# Count present days
present_count = 0
for status in attendance:
    if status== "P":
        present_count+= 1
# Total days
total_days= len(attendance)
# Calculate percentage
percentage =(present_count / total_days)*100
print("Attendance Percentage:", percentage)
```

# Problem 7: Count Character Frequency

```python
# Given string
text = "pythonp"
# Initialize empty dictionary
char_count = {}
# Loop through each character
for ch in text:
    if ch in char_count:
        char_count[ch]+= 1
    else:
        char_count[ch]= 1
print(char_count)
```

# Problem 9: Check if a Key Exists in a Dictionary
