# Problem 1: Social Media – Post Engagement Analyzer

```python
def analyze_engagement(like_list):
  likes=0
  for like in like_list:
    likes+=like

  if likes>=1000:
    analysis = "Viral Post"
  else:
    analysis = "Normal Engagement"
  print("Total Likes:", likes)
  print("Post Status:", analysis)

like_list= [200,150,300,250,150]
analyze_engagement(like_list)
```

# Problem 3: Agriculture – Rainfall Adequacy Checker


```python
def analyze_rainfall(rainfall_list, required_level):
    total =0
    for rainfall in rainfall_list:
      total+=rainfall
    average= total/len(rainfall_list)
    if average>=required_level:
        analysis= "Adequate Rainfall"
    else:
        analysis= "Inadequate Rainfall"
    print("Average Rainfall:", int(average))
    print("Rainfall Status:", analysis)

rainfall_data = [60,75,80,70,75]
required_level = 70
analyze_rainfall(rainfall_data, required_level)
```

# Problem 4: Social Media – Duplicate Account Detection

```python
def duplicate_accounts(usernames):
  unique_users= set()
  duplicate_found= False
  for user in usernames:
      if user in unique_users:
          duplicate_found= True
          break
      else:
          unique_users.add(user)
  if duplicate_found:
        print("Duplicate Accounts Found: Yes")
  else:
      print("Duplicate Accounts Found: No")

users= ["anish01","rahul22","john69","anish01"]
duplicate_accounts(users)
```

# Problem 6: Agriculture – Premium Crop Price Filte

```python
def filter_premium_crops(prices):
  # Used list comprehension to filter the premium crops
  premium_crops = [price for price in prices if price > 2000]
  print(f"Premium Crops: {premium_crops}")

crop_prices = [1500, 2500, 1800, 3200, 1900]
filter_premium_crops(crop_prices)
```

# Problem 9: E-Learning – Student Attendance Eligibility System

```python
def check_exam_eligibility(attendance_list):
    total_classes = len(attendance_list)
    classes = 0
    for status in attendance_list:
        if status == 1:
          classes += 1
    percent= (classes/total_classes) * 100
    if percent >= 75:
        eligible= "Eligible"
    else:
        eligible= "Not Eligible"
    print("Attendance Percentage:", percent)
    print("Exam Eligibility:", eligible)
attendance = [1, 1, 0, 1, 1]
check_exam_eligibility(attendance)
```
